#### Average Income Of People Their Buying Product

```python
price_data = pd.DataFrame({"Product":["KP281","KP481","KP781"],"Price":[1500,1750,2500]})

aero_new = pd.merge(aero,price_data,on = "Product")
aero_new.groupby("Product")["Income"].mean().reset_index()
```

#### No of Product Buying Age-Wise Category

```python

bin_ = [15,20,31,41,51]
label = ["Teenage","Adult","Middle Age","Old Age"]
aero_new["Age-Wise Category"] = pd.cut(aero_new["Age"],bins = bin_,labels = label)

aero_new["Age-Wise Category"].value_counts().reset_index().rename(columns = {"count":"No Of Count"})
```

#### No of Product Buying  Gender and Age-Wise Category 

```python
aero_new[["Gender","Age-Wise Category"]].value_counts().reset_index().rename(columns = {"count":"No Of Count"})
```

#### No of Product Buying MaritalStatus and Age-Wise Category

```python
aero_new[["MaritalStatus","Age-Wise Category"]].value_counts().reset_index().rename(columns = {"count":"No Of Count"})
```

#### Age-Wise Category people average Income

```python
aero_new.groupby("Age-Wise Category")["Income"].mean().reset_index().rename(columns = {"Income":"Avg Income"})
```

## Probability and statistics

```python
value = pd.crosstab(aero["Gender"],aero["MaritalStatus"],margins = True)
value.index.name = None
value.columns.name = None
value
```

## Marginal Probabllity

```python
# Probability Of P(Male) 
print("P(Male) = ",round(value.loc["Male"]["All"]/value.loc["All"]["All"]*100,2))
# Probability Of P(Female)
print("P(Female) = ",round(value.loc["Female"]["All"]/value.loc["All"]["All"]*100,2))
# Total Probability
print("Total Probability = ",round(value.loc["All"]["All"]/value.loc["All"]["All"]*100,2))
```

```python
# Probability Of P(Single) 
print("P(Single) = ",round(value["Single"]["All"]/value.loc["All"]["All"]*100,2))
# Probability Of P(Partnered)
print("P(Partnered) = ",round(value["Partnered"]["All"]/value.loc["All"]["All"]*100,2))
# Total Probability
print("Total Probability = ",round(value.loc["All"]["All"]/value.loc["All"]["All"]*100,2))
```

## Joint Probability

```python
# Probability Of P(Partnered ∩ Male)
print("P(Partnered ∩ Male) = ",round(value.loc["Male"]["Partnered"]/value.loc["All"]["All"]*100,2))
# Probability Of P(Partnered ∩ Female)
print("P(Partnered ∩ Female) = ",round(value.loc["Female"]["Partnered"]/value.loc["All"]["All"]*100,2))
# Total Probability
print("Total Probability Partnered = ",round(value["Partnered"]["All"]/value.loc["All"]["All"]*100,2))
```

```python
# Probability Of P(Single ∩ Male)
print("P(Partnered ∩ Male) = ",round(value.loc["Male"]["Single"]/value.loc["All"]["All"]*100,2))
# Probability Of P(Single ∩ Female)
print("P(Partnered ∩ Female) = ",round(value.loc["Female"]["Single"]/value.loc["All"]["All"]*100,2))
# Total Probability Single
print("Total Probability Single = ",round(value["Single"]["All"]/value.loc["All"]["All"]*100,2))
```

## Conditional Probability

```python
# Probability Of P(Partnered|Female)
print("P(Partnered|Female) = ",round(value.loc["Female"]["Single"]/value.loc["Female"]["All"]*100,2))
print("P(Partnered|Female) = ",round(value.loc["Female"]["Partnered"]/value.loc["Female"]["All"]*100,2))
```

```python
# Probability Of P(Partnered|Male)
print("P(Partnered|Male) = ",round(value.loc["Male"]["Single"]/value.loc["Male"]["All"]*100,2))
# Probability Of P(Partnered|Male)
print("P(Partnered|Male) = ",round(value.loc["Male"]["Partnered"]/value.loc["Male"]["All"]*100,2))
```
